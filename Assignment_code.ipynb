```python
from sentence_transformers import SentenceTransformer
# model = SentenceTransformer('bert-base-nli-mean-tokens')
model = SentenceTransformer('distiluse-base-multilingual-cased-v1')
```

```python
sentences=["this is an amazing app for online classes","good app for conducting online classes","app is user-friendly","good for video conferencing","unable to download zoom app",
           "my cat is good","want to download the app","app is not working"]
sentence_embeddings = model.encode(sentences)

sentence_embeddings.shape
```

```python
sentence_embeddings[1:].shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
arr=cosine_similarity(sentence_embeddings[0],sentence_embeddings[1].reshape(-1,1))
arr
```

```python
for i in range(1,len(sentences)):
  print(f"{sentences[i]} : {arr[0][i-1]}")
```
